## Data cleaning

### Stores

#### Nulos de CompetitionDistance en df_store

En primera instancia los eliminariamos al ser pocos, pero primero veamos si esa store tiene muchos registros de ventas en train

```python
stores_sin_competitiondistance = df_store[df_store["CompetitionDistance"].isna()]["Store"]
```

```python
stores_sin_competitiondistance
```

Registros de train con los store que no tienen CompetitionDistance

```python
len(df_train[df_train["Store"].isin(stores_sin_competitiondistance)])
```

Solo son 2642, dentro de más de 1 millón esto son pocos, asi que los eliminaremos

```python
df_store = df_store.dropna(subset=["CompetitionDistance"])
```

```python
df_store.info()
```

#### Nulos en CompetitionOpenSinceMonth y CompetitionOpenSinceYear

Estos nulos son falta de información, por lo que habría que imputar con mediana ya que hacerlo con 0's diría que la competición nunca abrió. Pero si imputamos con mediana lo que va a ocurrir es que utilizaremos datos que irán a "test", por lo que estaríamos creando un pequeño data leakage, así que dejaremos esa imputación para el apartado de feature engineering.

#### Nulos en Promo2SinceWeek y Promo2SinceYear

Estos nulos quieren decir que no hay Promo2 en ese local, asi que lo rellenaremos con 0's

```python
df_store["Promo2SinceWeek"] = df_store["Promo2SinceWeek"].fillna(0)
df_store["Promo2SinceYear"] = df_store["Promo2SinceYear"].fillna(0)
```

```python
df_store.info()
```

#### Nulos de PromoInterval

Veamos que categorías tiene el PromoInterval

```python
df_store["PromoInterval"].value_counts()
```

Solo tiene 3, asi que podríamos luego convertirla en dummies sin hacer tan complejo el dataset, completamos con "NoPromo" indicando que no existe promoción

```python
df_store["PromoInterval"] = df_store["PromoInterval"].fillna("NoPromo")
```

```python
df_store.info()
```

#### Hay tiendas duplicadas?

```python
df_store["Store"].duplicated().sum()
```

### Train

#### Convertir Date a tipo datetime

```python
df_train["Date"] = pd.to_datetime(df_train["Date"])
```

```python
df_train.info()
```

#### StateHoliday es object

Veamos porqué. Primero revisar que categorías tiene

```python
df_train["StateHoliday"].unique()
```

Existen categorías como "0", "a", 0

Entonces sabemos que "0" y 0 significan lo mismo, no hay holiday, asi que convertiremos todo en un string, así el 0, pasa a ser "0"

```python
df_train["StateHoliday"] = df_train["StateHoliday"].astype(str)
```

```python
df_train.info()
```
