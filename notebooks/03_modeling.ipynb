# Machine Learning

## Imports

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_percentage_error
```

## Importar dataframe

```python
DATA_PROCESSED_ROUTE = r"../data/processed/df_processed.csv"
```

```python
df = pd.read_csv(DATA_PROCESSED_ROUTE, dtype={"StateHoliday": str})
```

## Separar el DF en train y test

```python
df_train = df[
    df["Date"] < "2015-01-01"
]
```

```python
df_test = df[
    df["Date"] >= "2015-01-01"
]
```

## Eliminar Date

Los modelos no pueden interpretar fechas

```python
df_train = df_train.drop(columns=["Date"])
```

```python
df_test = df_test.drop(columns=["Date"])
```

## Separar X_train, X_test, y_train, y_test

```python
X_train = df_train.drop(columns=["Sales"])
y_train = df_train["Sales"]
```

```python
X_test = df_test.drop(columns=["Sales"])
y_test = df_test["Sales"]
```

## Obtener dummies

```python
X_train = pd.get_dummies(X_train)
```

```python
X_test = pd.get_dummies(X_test)
```

## Alinear

```python
X_train, X_test = X_train.align(X_test, join="left", axis=1, fill_value=0)
```

## Modelado

### Linear Regression V1

Instanciar

```python
model_linear_regression_v1 = LinearRegression()
```

Entrenar

```python
model_linear_regression_v1.fit(X_train, y_train)
```

Predecir (tanto para train y test)

```python
y_pred_linear_regresion_v1_train = model_linear_regression_v1.predict(X_train)
```

```python
y_pred_linear_regresion_v1_test = model_linear_regression_v1.predict(X_test)
```

Evaluar (de forma porcentual)

Evaluamos asi porque si la prediccion es 100 y el valor real 200, es un gran error, pero si la prediccion es 1000 y el real 1100 no lo es tanto

También evaluaremos el aprendizaje en train, para detectar overfitting

```python
mape_linear_regression_v1_train = mean_absolute_percentage_error(y_train, y_pred_linear_regresion_v1_train)
```

```python
mape_linear_regression_v1_test = mean_absolute_percentage_error(y_test, y_pred_linear_regresion_v1_test)
```

```python
print(f"Mape LinearRegression V1 train: {mape_linear_regression_v1_train * 100:.2f}%")
```

```python
print(f"Mape LinearRegression V1 test: {mape_linear_regression_v1_test * 100:.2f}%")
```
